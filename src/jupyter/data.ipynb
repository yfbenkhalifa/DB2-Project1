```python
import rdflib
import os
from utilities import Utilities as utils
import pandas
dataDir = '../../data/'
print(os.listdir(dataDir))
```

# Load Data


First we want to create the nodes for all continents and countries we want to consider

```python
# load dataset from csv
europe_dataset = dataDir + 'EnergyConsumption/EC_Europe.csv'
dfEurope = utils.creteDataframe(europe_dataset)
naDataset = dataDir + 'EnergyConsumption/EC_NA.csv'
countryCodes = dataDir + 'country_codes.csv'
dfNa = utils.creteDataframe(naDataset)
countryCodes = dataDir + 'country-codes.csv'
dfCountryCodes = utils.creteDataframe(countryCodes)
```

# Creating RDF

```python
# Load the required libraries
from rdflib import Graph, Literal, RDF, URIRef, Namespace
# rdflib knows about some namespaces, like FOAF
from rdflib.namespace import FOAF, XSD
# Create Namespaces
CNS =  Namespace("http://eulersharp.sourceforge.net/2003/03swap/countries#")
namespaces = {
    "country" : CNS,
    "foaf" : FOAF,
    "xsd" : XSD
}

# Create RDF graph
g = Graph()
# bind namespaces
utils.bindNamespaces(g, namespaces)
# Create country Nodes
utils.createNodes(g, dfCountryCodes, "FIFA", CNS.Country,CNS)
# Print nodes
g.print()

```

```python
# Print Turtle
print("--- saving serialization ---")
savePath = '../../data/energy.ttl'
with open(savePath + 'movies.ttl', 'w') as file:
    file.write(g.serialize(format='turtle'))
```
